```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years,runs)
# plt.plot(runs , years)

plt.xlabel("Year")
plt.ylabel("runs")
plt.title("sachin scored runs yearly")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli,label="virat kohli")
plt.plot(years,sehwag,label="virendra sehwag")

plt.xlabel("year")
plt.ylabel("runs scored")
plt.title("performance comparision")
plt.legend()
```

```python
plt.plot(years,kohli,"ro--",label="kohli")
plt.plot(years,sehwag,"g>:",label="sehwag")

plt.xlabel("year")
plt.ylabel("runs scored")
plt.title("performance comparision")
plt.grid(True)
plt.legend()
```

```python
plt.plot(years,kohli,color="orange",linestyle="--",label="kohli")
plt.plot(years,sehwag,color="green",linestyle="-.",label="sehwag")
plt.plot(years,runs,"g<:",label="tendulkar")


plt.xlabel("year")
plt.ylabel("runs scored")
plt.title("performance comparision")
plt.grid(True)
plt.legend()
```

```python
import matplotlib.pyplot as plt
import math

# Generate values
x = list(range(1, 11))  # integers 1 to 10
y = [math.factorial(i) for i in x]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', linewidth=2)

# Add labels and title
plt.title("Factorials Visualization")
plt.xlabel("n")
plt.ylabel("n!")

# Show grid
plt.grid(True, linestyle="--", alpha=0.7)
```
