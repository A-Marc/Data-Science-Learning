```python
data = {
"name": ["Alice", "Bob", "Charlie"],
"age": [25, 30, 35],
"city": ["Delhi", "Mumbai", "Bangalore"]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.index = ["a", "b", "c"]
df.columns = ["Name", "Age", "City"]
```

```python
print(df)
```
