```python
from sklearn import tree

# Step 1: Collect training data
features = [[140, 1], [130, 1], [150, 0], [170, 0]]  # [weight, texture]
labels = [0, 0, 1, 1]  # 0 = Apple, 1 = Orange

# Step 2: Train the model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, labels)

# Step 3: Make a prediction
print(clf.predict([[160, 0]]))  
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.figure(figsize=(6, 4))   # smaller figure
tree.plot_tree(
    clf,
    feature_names=["weight", "texture"],
    class_names=["Apple", "Orange"],
    filled=True,
    rounded=True,
    fontsize=8,          # smaller text
    proportion=True      # keeps nodes nicely scaled
)

plt.tight_layout()       # prevents overflow
plt.show()
```
